```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([
               [1,1],
               [1,2],
               [2,2],
               [2,3]
])
```

```python
y = np.dot(x, np.array([1,2])) + 3
```

```python
y
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.predict(np.array([[3,5]]))
```
